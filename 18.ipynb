```python
def is_subsequence(s_prime, s):
    it = iter(s)
    return all(event in it for event in s_prime)

tests = [
    (["Yahoo", "eBay", "Yahoo", "Oracle"],
     ["Amazon", "Yahoo", "eBay", "Yahoo", "Yahoo", "Oracle"],
     True),

    (["Yahoo", "Oracle", "eBay"],
     ["Amazon", "Yahoo", "eBay", "Yahoo", "Yahoo", "Oracle"],
     False),

    (["Amazon"],
     ["Amazon", "Yahoo", "eBay"],
     True),

    (["Oracle"],
     ["Amazon", "Yahoo", "eBay"],
     False),

    ([],
     ["Amazon", "Yahoo", "eBay"],
     True),

    (["Yahoo", "Yahoo", "Yahoo"],
     ["Yahoo", "Amazon", "Yahoo", "eBay", "Yahoo"],
     True),

    (["Yahoo", "Yahoo", "Yahoo", "Yahoo"],
     ["Yahoo", "Amazon", "Yahoo", "eBay", "Yahoo"],
     False),

    (["Amazon", "Yahoo", "eBay", "Yahoo", "Yahoo", "Oracle"],
     ["Amazon", "Yahoo", "eBay", "Yahoo", "Yahoo", "Oracle"],
     True),

    (["Amazon", "Oracle"],
     ["Oracle", "Amazon"],
     False),
]

for s_prime, s, expected in tests:
    result = is_subsequence(s_prime, s)
    status = "OK" if result == expected else "FAIL"
    print(f"[{status}] s'={s_prime} -> {result}")
```
