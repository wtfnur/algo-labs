```python
def place_base_stations(houses):
    if not houses:
        return []

    houses = sorted(houses)
    stations = []
    i = 0
    n = len(houses)

    while i < n:
        station = houses[i] + 4
        stations.append(station)

        while i < n and abs(houses[i] - station) <= 4:
            i += 1

    return stations

houses = [1, 2, 5, 7, 12]
print(place_base_stations(houses))
```
